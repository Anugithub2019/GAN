```python
DATA = "../GAN_exp_1000_epoch_500"
```

```python
FID_DATA = "CAT_FID_score.csv_2000"
```

```python
def cat(f):
     with open(f) as f:
         print(f.read())
```

```python
cat(f"{DATA}/intrinsic_dim_real.txt")
```

```python
#ChatGPT generated code block to extract the intrinsic dims from the above file

import re

def extract_intrinsic_dimension(input_string):
    # Define a regular expression pattern to match the intrinsic dimension
    pattern = r'Intrinsic Dimension of real data:\s*([\d.]+)'

    # Use re.search to find the match in the input string
    match = re.search(pattern, input_string)

    if match:
        # Extract the intrinsic dimension from the matched group
        intrinsic_dimension = float(match.group(1))
        return intrinsic_dimension
    else:
        # Return None if the pattern is not found
        return None

# Example usage
input_string = "Intrinsic Dimension of real data: 23.442501263446754 Standard error: 0.0"
real_intrinsic_dim = extract_intrinsic_dimension(input_string)

print("Intrinsic Dimension:", real_intrinsic_dim)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file using pandas
df = pd.read_csv(f'{DATA}/intrinsic_dim.csv')

# Create a plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Plot the data for intrinsic dimensions of generated data
sns.lineplot(x='epoch', y='intrinsic_dim', data=df, marker='o', linestyle='-', label='Intrinsic Dimensions of Generated Data', color='b')

# Add a horizontal line for intrinsic dimensions of real data
plt.axhline(y=real_intrinsic_dim, color='r', linestyle='--', label=f'Intrinsic Dimensions of Real Data ({real_intrinsic_dim:.2f})')

# Set plot labels and title
plt.xlabel('Epoch')
plt.ylabel('Intrinsic Dimensions')
#plt.title('Intrinsic Dimension as a Function of Epoch')

plt.xlim(0, 500)
plt.ylim(0, 30)

# Add a legend
plt.legend(loc='lower right')

# Show the plot
plt.show()
```
